##Building Block of Deep Neural Networks?

The building block of the deep neural networks is called the sigmoid neuron. Sigmoid neurons are similar to perceptrons, but they are slightly modified such that the output from the sigmoid neuron is much smoother than the step functional output from perceptron.

##Why to use Sigmoid Neuron Models?
* They can take Real Input and Output Data
* It is a non-linearly separable function and the output is smooth as we are using the Sigmoid Function
* We can do both Linear and Logistic Regression Tasks with this Model

## Mathematical Representation of Sigmoid Function:

The simplest formula for Signmod Neuron can be written as, 

$\Large {y = \frac{1}{1+e^{-(wx + b)}}}$

When you have only one feature, you can write the same formula as:

$\Large {y = \frac{1}{1 + e^{-(w_1x_1 + b)}}}$

When you have two features, then the same formula can be written as:

$\Large {y = \frac{1}{1 + e^{-(w_1x_1 + w_2x_2 + b)}}}$

Similarly, for n no. of features the signmod function can be written as:

$\Large {y = \frac{1}{1 + e^{-(\sum{w_ix_i} + b)}}}$

### Import Libaries

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns; sns.set()
from sklearn.metrics import mean_squared_error
from matplotlib.colors import LinearSegmentedColormap

%matplotlib inline
```

### 2D Plot of Sigmoid Function

If we want to visualize the 2D plot of Sigmoid Function, then we should have only one input feature and only one weight parameter associated to that weight.

```
def sigmoid(x, w, b):
  return 1 / (1 + np.exp(-(w*x + b)))
```

```
w = 1 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 0 #@param{type:'slider', min:-10, max:10, step:0.1}

X = np.linspace(-20, 20, 200)
Y = sigmoid(X, w, b)

plt.plot(X, Y)
plt.title('2-D plot of Sigmoid Neuron Function, w='+ str(w) + ' b='+str(b))
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(0,1)
plt.show()
```

Let's change the value of w and b but we will keep the input X same to see how the output curve looks like.

```
w = 0.2 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 0 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid(X, w, b)

plt.plot(X, Y)
plt.title('2-D plot of Sigmoid Neuron Function, w='+ str(w) + ' b='+str(b))
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(0,1)
plt.show()
```

```
w = 0 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 0 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid(X, w, b)

plt.plot(X, Y)
plt.title('2-D plot of Sigmoid Neuron Function, w='+ str(w) + ' b='+str(b))
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(0,1)
plt.show()
```

Now we will use the 2D Sigmoid Plot to see how we can use it to find the relation b/w x and y

```
X_input = np.array([2, 3, 4, 5, 8])
Y_input = np.array([0.047, 0.268, 0.73, 0.952, 0.999])

plt.plot(X_input, Y_input, '.', color='r')
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.ylim(0,1)
plt.show()
```

```
X = np.linspace(-10, 10, 100)
```

```
w = 2 #@param{type:'slider', min:-10, max:10, step:0.1}
b = -7 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid(X, w, b)

plt.plot(X_input, Y_input, '.', color='r')
plt.plot(X, Y)
plt.ylim(0, 1)
plt.show()
```

```
Y_pred = sigmoid(X_input, w, b)
mean_squared_error(Y_pred, Y_input)
```

### 3D Plot of Sigmoid Function

```
def sigmoid3D(x1, x2, w1, w2, b):
  return 1 / (1 + np.exp(-(w1*x1 + w2*x2 + b)))
```

```
X1 = np.linspace(-10, 10, 100)
X2 = np.linspace(-10, 10, 100)

XX1, XX2 = np.meshgrid(X1, X2)
```

```
my_cmap = LinearSegmentedColormap.from_list('', ['red', 'yellow', 'green'])
sns.reset_orig() # To reset back to matplotlib, otherwise the 3D plot will be lighter in shade
```

```
w1 = 0.5 #@param{type:'slider', min:-10, max:10, step:0.1}
w2 = 0 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 0 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid3D(XX1, XX2, w1, w2, b)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1, XX2, Y, cmap=my_cmap)
ax.set_title('3D Sigmoid Plot, w1='+ str(w1) + ', w2='+ str(w2) + ', b='+str(b))
ax.set_xlabel('XX1')
ax.set_ylabel('XX2')
ax.set_zlabel('Y')
ax.view_init(30, 140)
plt.show()
```

Let's change the w1, w2 and b value to visualize the change in output

```
w1 = 1 #@param{type:'slider', min:-10, max:10, step:0.1}
w2 = -0.7 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 0 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid3D(XX1, XX2, w1, w2, b)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1, XX2, Y, cmap=my_cmap)
ax.set_title('3D Sigmoid Plot, w1='+ str(w1) + ', w2='+ str(w2) + ', b='+str(b))
ax.set_xlabel('XX1')
ax.set_ylabel('XX2')
ax.set_zlabel('Y')
ax.view_init(30, 140)
plt.show()
```

```
w1 = 3.4 #@param{type:'slider', min:-10, max:10, step:0.1}
w2 = -1 #@param{type:'slider', min:-10, max:10, step:0.1}
b = 5.5 #@param{type:'slider', min:-10, max:10, step:0.1}

Y = sigmoid3D(XX1, XX2, w1, w2, b)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1, XX2, Y, cmap=my_cmap)
ax.set_title('3D Sigmoid Plot, w1='+ str(w1) + ', w2='+ str(w2) + ', b='+str(b))
ax.set_xlabel('XX1')
ax.set_ylabel('XX2')
ax.set_zlabel('Y')
ax.view_init(30, 140)
plt.show()
```
